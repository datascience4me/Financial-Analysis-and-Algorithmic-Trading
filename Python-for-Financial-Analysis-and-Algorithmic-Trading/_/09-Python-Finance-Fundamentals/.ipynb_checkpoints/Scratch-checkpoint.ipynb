```python
import pandas as pd
import quandl
```

```python
start = pd.to_datetime('2012-01-01')
end = pd.to_datetime('2017-01-01')
```

```python
apple = quandl.get('WIKI/AAPL.11', start_date=start, end_date=end)
cisco = quandl.get('WIKI/CSCO.11', start_date=start, end_date=end)
ibm = quandl.get('WIKI/IBM.11', start_date=start, end_date=end)
amazon = quandl.get('WIKI/AMZN.11', start_date=start, end_date=end)
```

```python
for df in [apple, cisco, ibm, amazon]:
    display(df.head())
```

```python
apple.iloc[0]['Adj. Close']
```

```python
for stock_df in (apple, cisco, ibm, amazon):
    stock_df['Normed Return'] = stock_df['Adj. Close']/stock_df.iloc[0]['Adj. Close']
```

```python
for df in [apple, cisco, ibm, amazon]:
    display(df.head(1))
    display(df.tail(1))
    print("----")
```

```python
# 30% in apple
# 20% in cisco
# 40% in amazon
# 10% in IBM
```

```python
# check zip code for first five rows
list(zip((apple.head(), cisco.head(), ibm.head(), amazon.head()), [.3, .2, .4, .1]))
```

```python
for stock_df, allo in zip((apple, cisco, ibm, amazon), [.3, .2, .4, .1]):
    stock_df['Allocation'] = stock_df['Normed Return']*allo
```

```python
for df in [apple, cisco, ibm, amazon]:
    display(df.head(1))
    display(df.tail(1))
    print("----")
```

```python
# invest 100000
#  [.3, .2, .4, .1] to (apple, cisco, ibm, amazon) respectively
for stock_df in (apple, cisco, ibm, amazon):
    stock_df['Position Values'] = stock_df['Allocation'] * 1000000
```

```python
for df in [apple, cisco, ibm, amazon]:
    display(df.head(1))
    display(df.tail(1))
    print("----")
```

```python
portfolio_val = pd.concat([apple['Position Values'],cisco['Position Values'],ibm['Position Values'],amazon['Position Values']],axis=1)
portfolio_val
```
