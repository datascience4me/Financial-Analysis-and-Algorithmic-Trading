# Testing for Stationarity

```python
df.head()
```

```python
# Augmented Dickey-Fuller unit root test
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(df['Milk in Pounds per Cow'])
```

```python
result
```

```python
def adf_check(time_series):
    result = adfuller(time_series)
    print("Augmented Dickey-Fuller unit root test")
    labels = ['ADF Test Statistic', 'p-value', '# of lags', 'Num of Obs. Used']
    
    for value, label in zip(result, labels):
        print(label+" : "+str(value))
        
    if result[1] <= 0.05:
        print("""\n\nstrong evidence against the null hypothesis, 
reject the null hypothesis. 
Data has no unit root and is stationary""")
    else:
        print("""\n\nweak evidence against null hypothesis, 
fail to reject null hypothesis,
time series has a unit root, 
indicating it is non-stationary """)
```

```python
adf_check(df['Milk in Pounds per Cow'])
```

```python
df['First Difference'] = df["Milk in Pounds per Cow"] - df["Milk in Pounds per Cow"].shift(1)
```

```python
df['First Difference'].plot()
```

```python
adf_check(df['First Difference'].dropna())
```

```python
df['Milk Second Difference'] = df["First Difference"] - df["First Difference"].shift(1)
```

```python
adf_check(df['Milk Second Difference'].dropna())
```

```python
df['Milk Second Difference'].plot()
```

```python
df['Seasonal Difference'] = df['Milk in Pounds per Cow'] - df['Milk in Pounds per Cow'].shift(12)
```

```python
df['Seasonal Difference'].plot()
```

```python
adf_check(df['Seasonal Difference'].dropna())
```

```python
df['Seasonal First Difference'] = df["First Difference"] - df["First Difference"].shift(12)
```

```python
df['Seasonal First Difference'].plot()
```

```python
adf_check(df['Seasonal First Difference'].dropna())
```
