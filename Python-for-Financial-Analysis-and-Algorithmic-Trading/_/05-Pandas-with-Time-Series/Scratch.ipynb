# Time resampling

```python
df = pd.read_csv('time_data/walmart_stock.csv')
```

```python
df.head()
```

```python
type(df['Date'][0])
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
type(df['Date'][0])
```

```python
df.head()
```

```python
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
df.index
```

```python
df.resample(rule='A')
```

A = year<br>
http://pandas.pydata.org/pandas-docs/stable/timeseries.html#offset-aliases

```python
df.resample(rule='A').mean()
```

```python
# SAMPLE BY BUSINESS QUARTERS (considers Fridays)
df.resample(rule='BQ').mean()
```

```python
# get the max of the year
df.resample(rule='A').max()
```

```python
# OWN RULE
def first_day(entry):
    return entry[0]

df.resample('A').apply(first_day)
```

#### Visualizations

```python
df['Close'].resample('A').mean().plot(kind = 'bar')
```

```python
df['Close'].resample('M').mean().plot(kind = 'bar', figsize=(16, 6))
```

# Rolling and Expanding 

```python
df.head()
```

```python
df['Open'].plot(figsize=(16, 6))
```

```python
# Signature: df.rolling(window, min_periods=None, freq=None, center=False, win_type=None, on=None, axis=0, closed=None)
# window : Size of the moving window. This is the number of observations used for
#          calculating the statistic. Each window will be a fixed size.

df.rolling(7).mean().head(15)
```

### The first 6 rows are blank because it takes the mean for the first rows, it is aggregating the previous 7 days

```python
df['Open'].plot()
df.rolling(window=7).mean()['Close'].plot(figsize=(16, 6))
```

```python
# TAKE A LOOK AT 30 DAY ROLLING AVG.
df['Open'].plot()
df.rolling(window=30).mean()['Close'].plot(figsize=(16, 6))
```

```python
# TAKE A LOOK AT 90 DAY ROLLING AVG.
df['Open'].plot()
df.rolling(window=90).mean()['Close'].plot(figsize=(16, 6))
```

```python
df['Close 30 Day Moving Avg.'] = df['Close'].rolling(window=30).mean()
df[['Close', 'Close 30 Day Moving Avg.']].plot(figsize=(16, 6))
```

```python
df['Close'].expanding().mean().plot(figsize=(16, 6))
```

# Bolinger bands

```python
# Close 20 MA (Moving Avg.)
df['Close: 20 Day Mean'] = df['Close'].rolling(20).mean()

# Upper = 20MA + 2*std(20)
df['Upper'] = df['Close: 20 Day Mean'] + 2*(df['Close'].rolling(20).std())

# Lower = 20MA - 2*std(20)
df['Lower'] = df['Close: 20 Day Mean'] - 2*(df['Close'].rolling(20).std())

# close
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].plot(figsize=(16, 6))
```

```python
# use slicing for closer look
df[['Close', 'Close: 20 Day Mean', 'Upper', 'Lower']].head(200).plot(figsize=(16, 6))
```
