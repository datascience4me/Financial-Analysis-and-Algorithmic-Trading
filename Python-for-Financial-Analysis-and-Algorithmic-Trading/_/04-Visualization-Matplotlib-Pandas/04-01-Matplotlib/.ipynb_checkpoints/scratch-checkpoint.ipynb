```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
print(x, y)
```

```python
plt.plot(x, y, 'r') # r = red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
```

```python
# subplot(nrows, ncols, index, **kwargs)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g+-')
```

### Object Oriented Method

```python
empty_canvas = plt.figure()

# Add set of axes to figure
axes = empty_canvas.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
# Creates blank canvas
empty_canvas = plt.figure()

axes1 = empty_canvas.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = empty_canvas.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes
# 20% from the bottom left
# 50% up from the bottom
# 40% of the canvas width
# 30% of the canvas height
```

```python
# Creates blank canvas
empty_canvas = plt.figure()

axes1 = empty_canvas.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = empty_canvas.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

axes1.plot(x, y, 'y')

axes2.plot(y, x, 'c')
axes2.set_title('Smaller Plot')
```

# Subplots

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)

axes[1].plot(y, x)

plt.tight_layout()
```

## Fig size and DPI

```python
fig = plt.figure(figsize=(3, 2))

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize = (8, 2))

axes[0].plot(x, y)

axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
ax.set_title('Title')
ax.set_xlabel('X_Label')
ax.set_ylabel('y_Label')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label= 'x squared')
ax.plot(x, x**3, label = 'x cubed')
ax.legend(loc=0)
```

## Setting colors, linewidths, linetypes

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#ff8c00', linewidth=3, linestyle='--')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=1, ls='-', marker="o", markersize=10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=1, ls='-', marker="o", markersize=20, 
       markerfacecolor='orange', markeredgewidth=3, markeredgecolor='green')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, ls='--')
ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, x, x**3)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].grid(ls='--', lw=0.5)

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale('log')
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, x, x**3)
ax.text(3, 40, 'text inside plot', fontsize=20, color='red')
```
